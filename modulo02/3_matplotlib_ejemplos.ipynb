## Hacer una Figura

```python
# Crear una figura
fig = plt.figure(figsize=(10, 5))

# Solicitar, de una cuadr√≠cula de 1x1 gr√°ficos, el primer eje.
ax = fig.add_subplot(1, 1, 1)

# Graficar los tiempos como variable en el eje x y las temperaturas como variable en el eje y
ax.plot(times, temps)

# Agregar etiquetas al gr√°fico
ax.set_xlabel('Tiempo')
ax.set_ylabel('Temperatura (¬∞K)')

ax.set_title('Figura de Temperatura')

ax.set_ylim(280, 320)
```
