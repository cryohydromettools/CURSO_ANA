# Visualizar y analizar las salidas

En este notebook vamos a analizar las salidas del modelo PDD.

## Importamos la libreria

```python
import xarray as xr
```

## Leemos el arvhivo de salida

```python
filename = './positive-deg-day-model-1.0/PDD_noSD/AG_output.nc'
ds = xr.open_dataset(filename)
ds
```

## Análisis espacial

```python
ds.sel(time=slice('20160901', '20170831'))['SMB'].sum('time').where(ds.MASK==1).plot()
```

## Análisis del promedio anual

```python
ds.sel(time=slice('20160901', '20170831'))['SMB'].sum('time').where(ds.MASK==1).mean()
```
